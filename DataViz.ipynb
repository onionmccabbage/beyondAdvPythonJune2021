# Visualizing Data with Charts

```python
import numpy as np
import pandas as pd
PREVIOUS_MAX_ROWS = pd.options.display.max_rows
pd.options.display.max_rows = 20
np.random.seed(12345)
import matplotlib.pyplot as plt
import matplotlib
plt.rc('figure', figsize=(10, 6))
np.set_printoptions(precision=4, suppress=True)
```

```python
data =np.arange(10) # an array range
data
plt.plot(data)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
```

```python
plt.plot(np.random.randn(50).cumsum(), 'k--')
```

```python
_ = ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
_
```

```python
fig, axes = plt.subplots(2,2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
plt.subplots_adjust(wspace=0, hspace=0)
```

```python
data = np.random.randn(30).cumsum()
plt.plot(data, 'k--', label='Default')
plt.plot(data, 'k-', drawstyle='steps-post', label='post-step lines')
plt.legend(loc='best')
```

```python
from datetime import datetime
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
# get the external data
data = pd.read_csv('data_sources/spx.csv', index_col=0, parse_dates=True)
data.head() # this is a DataFrame
spx = data['SPX'] # extracting just one Series
spx
spx.plot(ax=ax, style='r--')
# we can annotate the plot
crisis_data=[(datetime(2007,10,11), 'Peak'),
             (datetime(2008,3,12), 'Bear Stearns Fail'),
             (datetime(2008,9,15), 'Lehman Bankrupt')]
for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date)+75),
               xytext=(date, spx.asof(date) + 225),
               arrowprops=dict(facecolor='cyan', headwidth=8, width=2,
                                headlength=4),
               horizontalalignment='left', verticalalignment='top')
# Zoom in on 2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])

ax.set_title('Important dates in the 2008-2009 financial crisis')
```
